```python
import pandas as pd

# Step 1: Read the CSV file

with open('data.csv', 'r', encoding='utf-8', errors='replace') as f:
    # Step 2: Read the CSV file into a pandas DataFrame
    data = pd.read_csv(f)

# Step 3: Check unique values in the "Select Section" column to ensure correct filtering
print(data['Select Section'].unique())

# Step 2: Split the data based on the values in the "Select Section" column
section_1 = data[data['Select Section'] == 'CS4100']
section_2 = data[data['Select Section'] == 'CS5100 (Sec A - MW, 2:50 - 4:30)']
section_3 = data[data['Select Section'] == 'CS5100 (Sec B - MWTh, 10:30 - 11:35)']

# Step 3 (Optional): Save the split data into new CSV files
section_1.to_csv('section_1.csv', index=False)
section_2.to_csv('section_2.csv', index=False)
section_3.to_csv('section_3.csv', index=False)

# Or, if you want to just see the splits:
print("Section 1:")
print(section_1)
print("\nSection 2:")
print(section_2)
print("\nSection 3:")
print(section_3)
```
